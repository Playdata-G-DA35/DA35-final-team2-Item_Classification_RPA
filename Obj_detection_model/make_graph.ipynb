```python
import cv2
import os
import numpy as np
import test_method
import pandas as pd
from matplotlib import pyplot as plt
```

```python
csv1 = pd.read_csv('exp_result/s_20000_3(batch_32_80epoch_4.2hr)/results.csv')
csv2 = pd.read_csv('exp_result/s_40000_(batch48_100epoch_10.4hr)/results.csv')
csv3 = pd.read_csv('exp_result/result_50000_50ep/results.csv')
```

```python
train_box_loss1 = csv1['      train/box_loss']
train_box_loss2 = csv2['      train/box_loss']
train_box_loss3 = csv3['      train/box_loss']

plt.plot(train_box_loss1, label = 'Model1')
plt.plot(train_box_loss2, label = 'Model2')
plt.plot(train_box_loss3, label = 'Model3')
plt.legend()
plt.xlabel('epochs')
plt.ylim([0.01, 0.06])
plt.ylabel('loss')
plt.title('Train set/Box loss')
plt.savefig('exp_result/Train_box_loss.jpg', bbox_inches = 'tight')
```

```python
val_box_loss1 = csv1['        val/box_loss']
val_box_loss2 = csv2['        val/box_loss']
val_box_loss3 = csv3['        val/box_loss']

plt.plot(val_box_loss1, label = 'Model1')
plt.plot(val_box_loss2, label = 'Model2')
plt.plot(val_box_loss3, label = 'Model3')
plt.legend()
plt.xlabel('epochs')
plt.ylim([0.01, 0.06])
plt.ylabel('loss')
plt.title('Validation set/Box loss')
plt.savefig('exp_result/Val_box_loss.jpg', bbox_inches = 'tight')
```

```python
train_cls_loss1 = csv1['      train/cls_loss']
train_cls_loss2 = csv2['      train/cls_loss']
train_cls_loss3 = csv3['      train/cls_loss']

plt.plot(train_cls_loss1, label = 'Model1')
plt.plot(train_cls_loss2, label = 'Model2')
plt.plot(train_cls_loss3, label = 'Model3')
plt.legend()
plt.xlabel('epochs')
plt.ylim([0.00, 0.03])
plt.ylabel('loss')
plt.title('Train set/Class loss')
plt.savefig('exp_result/Train_cls_loss.jpg', bbox_inches = 'tight')
```

```python
val_cls_loss1 = csv1['        val/cls_loss']
val_cls_loss2 = csv2['        val/cls_loss']
val_cls_loss3 = csv3['        val/cls_loss']

plt.plot(val_cls_loss1, label = 'Model1')
plt.plot(val_cls_loss2, label = 'Model2')
plt.plot(val_cls_loss3, label = 'Model3')
plt.legend()
plt.xlabel('epochs')
plt.ylim([0.0, 0.03])
plt.ylabel('loss')
plt.title('Validation set/Class loss')
plt.savefig('exp_result/Val_cls_loss.jpg', bbox_inches = 'tight')
```
