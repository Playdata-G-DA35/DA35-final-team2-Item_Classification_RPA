```python
import cv2
import os
import numpy as np
import test_method
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def calc_precision(path):
    bottom_path_list, _ = test_method.make_path_list(f'{path}/bottom')
    onepiece_path_list, _ = test_method.make_path_list(f'{path}/onepiece')
    top_path_list, _ = test_method.make_path_list(f'{path}/top')

    person_total_num_list = []
    person_true_num_list = []
    product_total_num_list = []
    product_true_num_list = []

    for path in bottom_path_list:
        os.makedirs(f'{path}/person/crops/Bottom', exist_ok= True)
        os.makedirs(f'{path}/product/crops/Bottom', exist_ok= True)

        person_total_num = len(os.listdir(f'{path}/person')) - 3
        #product_true_num = len(os.listdir(f'{path}/person/crops/Bottom'))
        if(os.path.exists(f'{path}/person/crops/Pants')):
            person_true_num = len(os.listdir(f'{path}/person/crops/Pants'))
        else:
            person_true_num = len(os.listdir(f'{path}/person/crops/Bottom'))
        product_total_num = len(os.listdir(f'{path}/product')) - 3
        #product_true_num = len(os.listdir(f'{path}/product/crops/Bottom'))
        if(os.path.exists(f'{path}/product/crops/Pants')):
            product_true_num = len(os.listdir(f'{path}/product/crops/Pants'))
        else:
            product_true_num = len(os.listdir(f'{path}/product/crops/Bottom'))

        person_total_num_list.append(person_total_num)
        person_true_num_list.append(person_true_num)
        product_total_num_list.append(product_total_num)
        product_true_num_list.append(product_true_num)

    for path in onepiece_path_list:
        os.makedirs(f'{path}/person/crops/One piece', exist_ok= True)
        os.makedirs(f'{path}/product/crops/One piece', exist_ok= True)

        person_total_num = len(os.listdir(f'{path}/person')) - 3
        person_true_num = len(os.listdir(f'{path}/person/crops/One piece'))
        product_total_num = len(os.listdir(f'{path}/product')) - 3
        product_true_num = len(os.listdir(f'{path}/product/crops/One piece'))

        person_total_num_list.append(person_total_num)
        person_true_num_list.append(person_true_num)
        product_total_num_list.append(product_total_num)
        product_true_num_list.append(product_true_num)

    for path in top_path_list:
        os.makedirs(f'{path}/person/crops/Top', exist_ok= True)
        os.makedirs(f'{path}/product/crops/Top', exist_ok= True)

        person_total_num = len(os.listdir(f'{path}/person')) - 3
        person_true_num = len(os.listdir(f'{path}/person/crops/Top'))
        product_total_num = len(os.listdir(f'{path}/product')) - 3
        product_true_num = len(os.listdir(f'{path}/product/crops/Top'))

        person_total_num_list.append(person_total_num)
        person_true_num_list.append(person_true_num)
        product_total_num_list.append(product_total_num)
        product_true_num_list.append(product_true_num)

    return person_total_num_list, person_true_num_list, product_total_num_list, product_true_num_list
```

```python
bottom_path = '../dataset/test_set/bottom'
onepiece_path = '../dataset/test_set/onepiece'
top_path = '../dataset/test_set/top'
```

```python
test_path = 'test_result/20000img_80epoch/20000img_80epoch_0.1conf'
```

```python
_, bottom_name_list = test_method.make_path_list(bottom_path)
_, onepiece_name_list = test_method.make_path_list(onepiece_path)
_, top_name_list = test_method.make_path_list(top_path)
```

```python
def make_confusion_csv(result_path, col_name):
    folder_list = os.listdir(result_path)
    
    path_list = []
    for folder in folder_list:
        path_list.append(f'{result_path}/{folder}')

    df_list = []
    df = pd.DataFrame(columns=col_name)
    
    # 폴더별 실행 -> 9번
    for idx in range(len(path_list)):
        prsn_total_list, prsn_true_list, prd_total_list, prd_true_list = calc_precision(path_list[idx])
        new_row = []
        # conf 값 추가
        new_row.append(0.1 + 0.1*idx)

        # 전체, 착샷, 제품샷, 정확도
        new_row.append(((sum(prsn_true_list) + sum(prd_true_list)) / (sum(prsn_total_list) + sum(prd_total_list)))*100)
        new_row.append((sum(prsn_true_list)/sum(prsn_total_list))*100)
        new_row.append((sum(prd_true_list)/sum(prd_total_list))*100)

        # prsn, prd 순으로 카테고리별 추가
        # 제품별 실행 -> 18번
        for idx in range(len(prsn_total_list)):
            new_row.append((prsn_true_list[idx]/prsn_total_list[idx])*100)
            new_row.append((prd_true_list[idx]/prd_total_list[idx])*100)
        
        # df에 추가
        df = pd.concat([df, pd.DataFrame([new_row], columns= col_name)], ignore_index=True)
        df.to_csv(f'{result_path}_acc.csv', index=False)

    return df
```

```python
col_name = ['conf', 'all_acc', 'prsn_acc', 'prd_acc']

for name in bottom_name_list:
    col_name.append(name + '_prsn')
    col_name.append(name + '_prd')

for name in onepiece_name_list:
    col_name.append(name + '_prsn')
    col_name.append(name + '_prd')

for name in top_name_list:
    col_name.append(name + '_prsn')
    col_name.append(name + '_prd')

print(col_name)
print(len(col_name))
```
